```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                     columns=iris['feature_names'] + ['target'])
print("Iris dataset loaded successfully. Here are the first 5 rows:")
print(iris_df.head())
```

```python
data = datasets.load_iris()
data.target_names
data.feature_names
```

```python
data.target
```

```python
x = data.data
x.shape
```

```python
y = data.target
y.shape
```

```python
y= y.reshape(-1,1)
y.shape
```

```python
plt.figure(figsize=(18, 8))
plt.scatter(x.T[0], x.T[2])
plt.title('Iris Dataset',fontsize= 20)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
lr = LogisticRegression()
iris_model = lr.fit(X_train, y_train)
predictions = iris_model.predict(X_test)
```

```python
plt.errorbar(range(1,len(y_test)+1), y_test.flatten(), yerr=np.abs(y_test.flatten() - predictions), fmt='^k', ecolor='red')
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```
